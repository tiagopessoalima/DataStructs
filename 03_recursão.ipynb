# Recursão

É um conceito fundamental na matemática e na programação, simplifica a resolução de problemas complexos ao permitir a definição de objetos e funções em termos de si mesmos. Para demonstrar sua aplicação prática, vamos explorar o cálculo do fatorial, começando por uma abordagem não recursiva e, em seguida, apresentando a elegante solução recursiva.





## Fatorial

O fatorial, representado por $n!$, é um conceito matemático que se aplica a números naturais inteiros não negativos. Ele é calculado pela multiplicação do número $n$ por todos os seus antecessores positivos, até chegar a $1$. O fatorial de $0$ é definido como $1$ $(0! = 1)$.

image.png

## Fatorial - Abordagem sem Recursividade

A função `fatorial_nao_recursivo(n)` calcula o fatorial de um número `n`  sem utilizar recursividade. Primeiramente, ela verifica se `n` é igual a $0$. Se for, retorna $1$, já que o fatorial de $0$ é definido como $1$. Caso contrário, a função inicializa uma variável `resultado` com o valor $1$. Em seguida, um loop for itera de $1$ até `n`, multiplicando `resultado` por cada número `i` no intervalo. Após o loop, a função retorna o valor final de `resultado`, que representa o fatorial calculado de `n`. Em essência, a função implementa um processo iterativo de multiplicação acumulada para calcular o fatorial, demonstrando uma abordagem não recursiva para esse problema matemático.

```python
def fatorial_nao_recursivo(n):
  """Calcula o fatorial de um número n de forma não recursiva."""
  if n == 0:
    return 1
  else:
    resultado = 1
    for i in range(1, n + 1):
      resultado *= i
    return resultado
```

Demonstração do cálculo do fatorial de 1 a 5:

```python
for i in range(1, 6):
  print(f"O fatorial de {i} é {fatorial_nao_recursivo(i)}")
```

## Fatorial - Abordagem com Recursividade



A função `fatorial_recursivo(n)` oferece uma solução elegante para o cálculo do fatorial, aproveitando o poder da recursividade. Primeiramente, assim como na versão não recursiva, ela verifica se `n` é igual a $0$. Se for, retorna $1$, representando o caso base do fatorial. Caso contrário, a função entra em ação recursiva. Ela retorna o produto de `n` pelo resultado da chamada `fatorial_recursivo(n - 1)`. Essa chamada recursiva calcula o fatorial de `n - 1`, e o resultado é então multiplicado por `n` para obter o fatorial de `n`. Em essência, a função se chama repetidamente, reduzindo o problema a cálculos de fatoriais de números menores, até atingir o caso base (0!), que interrompe a recursão e permite que os resultados sejam combinados de volta para calcular o fatorial original.

```python
def fatorial_recursivo(n):
  """Calcula o fatorial de um número n de forma recursiva."""
  if n == 0:
    return 1
  else:
    return n * fatorial_recursivo(n - 1)
```

Demonstração do cálculo do fatorial de 1 a 5:

```python
for i in range(1, 6):
  print(f"O fatorial de {i} é {fatorial_recursivo(i)}")
```

# Leis da Recursividade

Para que um algoritmo recursivo funcione corretamente e evite cair em um loop infinito, ele deve seguir três leis fundamentais:

- Um algoritmo recursivo deve **possuir um caso base** (*base case*).
- Um algoritmo recursivo deve modificar o seu estado e **se aproximar do caso base**.
- Um algoritmo recursivo **deve chamar a si mesmo**, recursivamente.

No código do fatorial recursivo, podemos identificar claramente essas três leis:

- Caso base: `if n == 0: return 1`
- Aproximação do caso base: `fatorial_recursivo(n - 1)` (o valor de n diminui a cada chamada)
- Chamada recursiva: `return n * fatorial_recursivo(n - 1)` (a função chama a si mesma)

Ao seguir essas leis, garantimos que a recursividade seja utilizada de forma controlada e eficiente, permitindo a resolução de problemas complexos de maneira elegante e concisa. Aprofundaremos nosso estudo da recursividade com a sequência de Fibonacci.

## **Sequência de Fibonacci**

Definiremos agora uma sequência matemática muito famosa chamada Sequência de Fibonacci: 0,1,1,2,3,5,8,13,…. Nessa sequência, os dois termos iniciais são 0 e 1, e cada termo subsequente é igual à soma dos dois termos anteriores. O cálculo do n-ésimo termo da sequência de Fibonacci pode ser definido recursivamente da seguinte forma:

\begin{align*}
	F_n&=
	\begin{cases}
		n & \mbox{ se } n=0 \mbox{ ou } n=1,\\
		F_{n-1}+F_{n-2}& \mbox{ caso contrário }.
	\end{cases}
\end{align*}

A partir da definição acima, podemos implementar o cálculo do n-ésimo termo da sequência assim:

```python
def fib_recursivo(n):
    """Calcula o n-ésimo termo da sequência de Fibonacci de forma recursiva."""
    if n <= 1:
        return n  # Casos base: 0 e 1
    else:
        return fib_recursivo(n - 1) + fib_recursivo(n - 2)
```

Exemplo de uso:

```python
fib_recursivo(5)
```

Existem formas mais eficientes de implementarmos a função `fib`, sem usar recursão, como a seguinte:

```python
def fib_nao_recursivo(n):
    """Calcula o n-ésimo termo da sequência de Fibonacci de forma não recursiva (iterativa)."""
    if n <= 1:
        return n
    fib_n_menos_2 = 0
    fib_n_menos_1 = 1
    for i in range(2, n):
        fib_n = fib_n_menos_1 + fib_n_menos_2
        fib_n_menos_2 = fib_n_menos_1
        fib_n_menos_1 = fib_n
    return fib_n_menos_1 + fib_n_menos_2
```

Exemplo de uso:

```python
fib_nao_recursivo(50)
```

### Fibonacci Recursivo vs. Não Recursivo

O código abaixo realiza a comparação de desempenho entre duas implementações de uma função que calcula o valor de Fibonacci: uma versão recursiva e outra não-recursiva. Ele define uma lista de valores de `n` (números para os quais a função de Fibonacci será calculada), que vai de $1$ até $45$ (não incluso), em incrementos de $5$. Para cada valor de n, o tempo de execução das duas funções é medido utilizando a função `time.time()`, que captura o tempo antes e após a execução da função, armazenando a diferença (tempo de execução) em duas listas separadas para as versões recursiva e não-recursiva. Por fim, os tempos de execução são plotados em um gráfico, onde o eixo x representa os diferentes valores de `n`, e o eixo y representa o tempo de execução correspondente. O gráfico compara visualmente os dois métodos, permitindo observar qual deles é mais eficiente à medida que `n` aumenta.

```python
import matplotlib.pyplot as plt
import time

# Valores de 'n' para teste e listas para armazenar tempos
valores_n = list(range(0, 45, 5))  # De 0 a 45 (não incluso), com incrementos de 5
tempos_recursivo = []
tempos_nao_recursivo = []

# Medição de tempo para cada função
for n in valores_n:
    start_time = time.time()
    fib_recursivo(n)
    tempos_recursivo.append(time.time() - start_time)

    start_time = time.time()
    fib_nao_recursivo(n)
    tempos_nao_recursivo.append(time.time() - start_time)

# Plot do gráfico
plt.figure(figsize=(10, 6))
plt.plot(valores_n, tempos_recursivo, label='Recursivo', marker='o')
plt.plot(valores_n, tempos_nao_recursivo, label='Não Recursivo', marker='x')
plt.title('Comparação de Tempo de Execução: Fibonacci Recursivo vs. Não Recursivo')
plt.xlabel('Valor de n')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```

O código abaixo calcula a razão entre os tempos de execução das versões recursiva e não-recursiva de Fibonacci para diferentes valores de `n`, e gera um gráfico de barras que exibe essas razões no eixo y e os valores de `n` no eixo x.



```python
# Calcula a razão entre os tempos de execução das versões recursiva e não recursiva
razoes_tempos = [t_rec / t_nao_rec for t_rec, t_nao_rec in zip(tempos_recursivo, tempos_nao_recursivo)]

# Criação do gráfico de barras
fig, ax = plt.subplots()
ax.bar(valores_n, razoes_tempos)

# Adiciona rótulos e título ao gráfico
ax.set_xlabel('Valor de n')
ax.set_ylabel('Razão entre Tempos (Recursivo / Não Recursivo)')
ax.set_title('Razão entre Tempos de Execução Fibonacci')

# Exibe o gráfico
plt.show()
```

A recursão pode parecer menos vantajosa ao comparar a função fib recursiva com sua versão iterativa. Afinal, a versão iterativa não é muito mais complicada que a versão recursiva correspondente, e a versão iterativa é mais eficiente. Na prática, a versão não recursiva de um algoritmo tende a ser mais eficiente que a versão recursiva. Entretanto, em alguns casos a versão não recursiva é tão mais complicada que acaba sendo mais vantajoso usar a versão recursiva (por questões de legibilidade e facilidade de manutenção).








## Exercícios

Outro exemplo consiste na definição de exponenciação. Como sabemos, se $x\neq 0$, tem-se:

$$
x^n = \begin{cases}
1 & \text{se } n = 0, \\
x \cdot x^{n-1} & \text{se } n > 0, \\
\frac{1}{x^{-n}} & \text{se } n < 0.
\end{cases}
$$

```python
def exponencial(x, n):
    # Caso base: qualquer número elevado a 0 é 1
    if n == 0:
        return 1
    # Se o expoente for negativo, calcula o inverso da base
    elif n < 0:
        return 1 / exponencial(x, -n)
    # Caso recursivo: x^n = x * x^(n-1)
    else:
        return x * exponencial(x, n - 1)

# Exemplo de uso
base = 2
expoente = 3
resultado = exponencial(base, expoente)
print(f"{base}^{expoente} = {resultado}")
```

Um exemplo clássico de algoritmo recursivo é o de Euclides, utilizado para calcular o máximo divisor comum (MDC) entre dois números inteiros. A ideia reside na propriedade de que o MDC de dois números é igual ao MDC do segundo número e do resto da divisão do primeiro pelo segundo, ou seja: `mdc(a, b) = mdc(b, a % b)`. A recursão se repete até que o resto seja zero, o que caracteriza o caso base.

```python
def mdc(a, b):
  # Caso base: Se `b` é 0, `a` é o MDC.
  if b == 0:
    return a
  # Caso recursivo: Reduz o problema para encontrar o MDC de números menores.
  else:
    return mdc(b, a % b)

# Exemplo de uso
num1 = 48
num2 = 18
resultado = mdc(num1, num2)
print(f"O MDC de {num1} e {num2} é {resultado}")
```
